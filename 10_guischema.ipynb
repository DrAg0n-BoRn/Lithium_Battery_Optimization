```python
from ml_tools.schema import FeatureSchema, create_guischema_template

from paths import PM
from helpers.constants import CONTINUOUS_FEATURES_RANGE, TARGETS
```

## 1. Load Feature Schema

```python
feature_schema = FeatureSchema.from_json(directory=PM.optimization_engineering)
```

## 2. Construct multibinary groups

```python
feature_schema.categorical_feature_names
```

```python
multibinary_groups = {
    "Dopant" : [col_name for col_name in feature_schema.categorical_feature_names if "Dopant_" in col_name],
    "Electrolyte Solvent" : [col_name for col_name in feature_schema.categorical_feature_names if "Electrolyte Solvent_" in col_name],
    "Precursor Method": [col_name for col_name in feature_schema.categorical_feature_names if "Precursor Method_" in col_name],
    "Space": [col_name for col_name in feature_schema.categorical_feature_names if "Space_" in col_name]
}
```

```python
# assert that all groups sum up to the total number of categorical features minus 5 (other categorical features)
assert sum(len(group) for group in multibinary_groups.values()) == len(feature_schema.categorical_feature_names) - 5, "Multibinary groups do not cover all categorical features correctly."
```

## 3. Create GUI Schema

```python
create_guischema_template(directory=PM.optimization_engineering,
                          feature_schema=feature_schema,
                          continuous_ranges=CONTINUOUS_FEATURES_RANGE,
                          targets=TARGETS,
                          multibinary_groups=multibinary_groups)
```
